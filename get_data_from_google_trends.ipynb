```python
from pytrends.request import TrendReq
import numpy as np
import pandas as pd
```

```python
# Login to Google. Only need to run this once, the rest of requests will use the same session.
pytrend = TrendReq()
```

```python
items = pd.read_csv('items.csv', usecols = ['item_code', 'item_name'])
item_name = items['item_name'].tolist()

kw_list1 = []
kw_list1.append(item_name[0])
kw_list1.append(item_name[1])
kw_list1.append(item_name[2])
kw_list1
```

```python
stores_geo = pd.read_csv('stores_geo.csv', usecols = ['store_code', 'geo_code'])
stores_geo = stores_geo.loc[stores_geo['store_code'] == 1]
geo_code = stores_geo['geo_code'].tolist()
geo_code
```

```python
# Create payload and capture API tokens. Only needed for interest_over_time(), interest_by_region() & related_queries()
kw_list1=['chicken drumstick', 'sweet potato', 'chocolate ice cream']
kw_list = []
df2 = pd.DataFrame()

for i in range(1, 4):
    kw_list = [kw_list1[i-1]]
    #pytrend.build_payload(kw_list, timeframe='2010-02-05 2012-10-30', geo=str(geo_code[0]), gprop='')
    pytrend.build_payload(kw_list, timeframe='2010-02-05 2012-10-30', geo='CA-AB', gprop='')
    
    # Interest Over Time
    interest_over_time_dict = pytrend.interest_over_time()    
    
    interest_over_time_df = pd.DataFrame.from_dict(interest_over_time_dict[str(kw_list1[i-1])].items())    
    df = pd.DataFrame({'Store' : 1,
                       'Item' : i,
                       'Date' : interest_over_time_df[0],
                       'Weekly_Demand' : interest_over_time_df[1]})

    df2 = df2.append(df)
    
df2.to_csv('interest_over_time_s1_1_3.csv', index = False)
```
